## Exercício 1 – Explorando o Vocabulário


Treine um modelo Word2Vec com o corpus fornecido:

- Liste o vocabulário
- Mostre o vetor da palavra "rei"
- Informe o tamanho do vetor


```python
sentences = [
    ["rei", "rainha", "homem", "mulher"],
    ["paris", "franca", "roma", "italia"],
    ["homem", "trabalha"],
    ["mulher", "trabalha"],
]
```

```python
%pip install gensim
```

```python
from gensim.models import Word2Vec, KeyedVectors
import numpy as np

model = Word2Vec(sentences=sentences, vector_size=10, window=2, min_count=1)

print(f"Lista do vocalubario {model.wv.index_to_key} \n")

vetor_rei = model.wv['rei']

print(f"tamanho do vetor de rei: {vetor_rei.shape} \n" )
print("Vetor para a palavra rei: \n", vetor_rei)
```

## Exercício 2 – Similaridade Semântica


- Calcule a similaridade entre:
  - "rei" e "rainha"
  - "rei" e "italia"
- Compare os valores


```python
similaridade_rei_rainha = model.wv.similarity("rei", "rainha")
similaridade_rei_italia = model.wv.similarity("rei", "italia")

print(f"Similaridade entre a palavra 'rei' e 'rainha': {similaridade_rei_rainha:.4f}")

print(f"Similaridade entre a palavra 'rei' e 'italia': {similaridade_rei_italia:.4f}")

print(f"A diferença de similaridade entre os 2 pares de palavras a cima é de: {similaridade_rei_rainha - similaridade_rei_italia:.4f}")
```

## Exercício 3 – Palavras Mais Próximas


- Encontre as 3 palavras mais similares a "homem"
- Encontre as 3 palavras mais similares a "paris"
- Explique por que os resultados fazem sentido (ou não)


```python
print("Palavras mais similares a homem:")
for palavra, score in model.wv.most_similar("homem", topn=3):
    print(f" - {palavra} : {score:.4f}")  
    
print("-"*50)
    
print("Palavras mais similares a paris:")
for palavra, score in model.wv.most_similar("paris", topn=3):
    print(f" - {palavra} : {score:.4f}")
```

### R: A maioria dos resultados não fazem sentido, pois o vocabulario dado não apresenta quase nenhuma relação entre as palavras, fazendo com que o calculo do word2vec seja totalmente impreciso

## Exercício 4 – Analogia Vetorial


Resolva a analogia:

"rei" está para "homem" assim como "rainha" está para ?

Use operações vetoriais.
Interprete o resultado.


```python
comparacao = model.wv.most_similar(positive=["rainha", "homem"], negative=["rei"], topn=1)
print(f"A analogia acima apresenta o resultado: {comparacao}")
```

### Interpretação: pelo baixo vocabulario dado, a palavra que o modelo conseguiu interpretar como resposta para a analogia é paris, ja que nos vetores das palavras, rainha deve ser o mais proximo seguindo a analogia de "rei esta para homem assim como rainha esta para?"
