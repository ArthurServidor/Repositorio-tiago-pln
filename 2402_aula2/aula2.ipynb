# EXEMPLO 1

```python
import spacy

nlp = spacy.load("pt_core_news_sm")
texto = "O professor Tiago explicou PLN usando spaCy na Fatec."

doc = nlp(texto)

for token in doc:
    print(f"Texto: {token.text:10} |"
          f"Lemma: {token.lemma_:10} |"
          f"POS: {token.pos_:6} |"
          f"Stop: {token.is_stop}"
    )

print("\nEntidades:")
for ent in doc.ents:
    print(ent.text, ent.label_)
```

### EX 3

Dado um texto qualquer a sua escolha, faça:

a) Extraia apenas os verbos

b) Extraia apenas os substantivos

c) Conte quantos verbos existem no texto

• Use token.pos_

• Desafio extra:

– Mostrar verbo + seu sujeito usando dependência sintática

```python
xd = nlp("Visualize e faça o download dos detalhes de crédito aqui")

vertQtd = 0

for token in xd:
    if(token.pos_ == "VERB"):
        vertQtd +=1
        print(f"{token.text} - {token.pos_}")
        
print("--------------------------")

for token in xd:
    if(token.pos_ == "NOUN"):
        print(f"{token.text} - {token.pos_}")
        
print("--------------------------")

print(f"Quantidade de verbos: {vertQtd}")
```

### EX 4

• Dado um parágrafo maior:

– Remova stopwords

– Remova pontuação

– Gere uma lista de palavras "limpas"

– Mostre as 5 palavras mais frequentes

• Objetivo:

– Introduzir pré-processamento para Machine Learning

```python
xd = nlp("A inteligência artificial está transformando a educação superior no Brasil. Machine Learning é fundamental para o futuro da tecnologia. Os algoritmos precisam ser eficientes e escaláveis para processar grandes volumes de dados.")

palavras_limpas = [token.text for token in xd if not token.is_stop and not token.is_punct]

print("=== PALAVRAS LIMPAS ===")
print(palavras_limpas)

from collections import Counter
frequencia = Counter(palavras_limpas)

print("=== 5 PALAVRAS MAIS FREQUENTES ===")
for palavra, freq in frequencia.most_common(5):
    print(f"{palavra}: {freq}")
```

### EX 5

• Crie um programa que:

a) Conte a frequência de cada palavra

b) Ordene da mais frequente para a menos frequente

c) Ignore:

• Stopwords

• Pontuação

• Palavras com menos de 3 letras

```python
xd = nlp("A inteligência artificial está transformando a educação superior no Brasil. Machine Learning é fundamental para o futuro da tecnologia. Os algoritmos precisam ser eficientes e escaláveis para processar grandes volumes de dados.")

palavras_filtradas = [token.text for token in xd if not token.is_stop and not token.is_punct and len(token.text) >= 3]

from collections import Counter
frequencia = Counter(palavras_filtradas)

print("FREQUÊNCIA DE PALAVRAS (ORDENADO)")
for palavra, freq in frequencia.most_common():
    print(f"{palavra}: {freq}")
```

## R: Para identificar mais rapido pessoas citadas, locais citados, datas citadas dentro de um documento gigante

```python
xd = nlp("A OpenAI foi fundada em 2015 e possui escritórios em São Francisco.")

print("Entidades")
for ent in xd.ents:
    print(f"{ent.text} -> {ent.label_}")

print("--------------------------")

print("Contagem de entidades:")
contagem = Counter(ent.label_ for ent in xd.ents)
for label, cnt in contagem.items():
    print(f"{label}: {cnt}")
```
