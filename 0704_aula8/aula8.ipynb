```python
import sounddevice as sd
from scipy.io.wavfile import write
import whisper
from datetime import datetime
```

# Função para gravar áudio

```python
def gravar_audio(nome_arquivo="audio.wav", duracao=5, fs=44100):
    print("Gravando áudio...")
    audio = sd.rec(int(duracao * fs), samplerate=fs, channels=1)
    sd.wait()
    write(nome_arquivo, fs, audio)
    print("Gravação finalizada")
```

### Carregando o modelo Whisper

| Modelo | Tamanho | Qualidade | Velocidade   |
| ------ | ------- | --------- | ------------ |
| tiny   | ~75 MB  | baixa     | muito rápido |
| base   | ~150 MB | boa       | rápido       |
| small  | ~500 MB | melhor    | médio        |
| medium | ~1.5 GB | alta      | mais lento   |
| large  | ~3 GB   | excelente | pesado       |

```python
print("Carregamento modelo...")
model = whisper.load_model("small")
print("Modelo carregado!")
```

# Função para transcrição

```python
def transcrever_audio(nome_arquivo="audio.wav"):
    print("Transcrevendo áudio...")
    result = model.transcribe(nome_arquivo, language='pt')
    texto = result["text"]
    print("Texto:", texto)
    return texto.lower()
```

# Classificador de comandos

```python
def classificar_comando(texto):
    if "luz" in texto and "ligar" in texto:
        print("Ligando a luz")
    
    elif "luz" in texto and "desligar" in texto:
        print(" Desligando a luz")
        
    elif "hora" in texto:
        print(" Agora são:", datetime.now().strftime("%H:%M"))
        
    elif "data" in texto:
        print(" Hoje é", datetime.now().strftime("%d/%m/%Y"))
        
    else:
        print("Comando não reconhecido")
```

# Pipeline completo

```python
def executar_pipeline():
    gravar_audio()
    texto = transcrever_audio()
    classificar_comando(texto)

executar_pipeline()
```
