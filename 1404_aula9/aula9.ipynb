```python
import sounddevice as sd
import scipy.io.wavfile as wav
import whisper
import pyttsx3
import time
```

```python
# ======================
# 1. GRAVAÇÃO DE ÁUDIO
# ======================

def gravar_audio(nome_arquivo="audio.wav", duracao=5, fs=44100):
    print("Preparando para gravar...")
    
    # contador regressivo
    for i in range(3, 0, -1):
        print(i)
        time.sleep(1)
        
    print("Gravando...")
    audio = sd.rec(int(duracao * fs), samplerate=fs, channels=1)
    sd.wait()
    
    wav.write(nome_arquivo, fs, audio)
    print("Áudio gravado!")
```

```python
# ===============================
# 2. TRANSCRIÇÃO (Whisper)
# ===============================

def transcrever_audio(arquivo):
    print("Transcrevendo...")
    modelo = whisper.load_model("base")

    resultado = modelo.transcribe(arquivo, language="pt")
    texto = resultado["text"]

    print("Texto reconhecido:", texto)
    return texto
```

```python
# =================================
# 3. GERAR RESPOSTA
# =================================

def gerar_resposta(texto):
    if "oi" in texto:
        return "Olá! Como posso ajudar você?"
    elif "nome" in texto:
        return "Eu sou um assistente virtual feito em Python."
    elif "hora" in texto:
        return time.strftime("Agora são %H horas e %M minutos.")
    else:
        return "Desculpe, ainda não sei responder isso."
```

```python
# ==================================
# 4. Texto para ÁUDIO (pyttsx3)
# ==================================
def sintetizar_voz(texto):
    print("Gerando fala...") 
    
    engine = pyttsx3.init()
    
    # Configurações de voz
    engine.setProperty('rate', 180) #velocidade da fala
    engine.setProperty('volume', 1.0) #volume (0.0 a 1.0)
    
    # Selecionar voz(varia por sistema)
    voices = engine.getProperty('voices')
    
    for voice in voices:
        if "brazil" in voice.name.lower() or "portuguese" in voice.name.lower():
            engine.setProperty('voice', voice.id)
            break
    
    #Fala o texto
    engine.say(texto)
    engine.runAndWait()
```

```python
# ==================================
# PIPELINE COMPLETO
# ==================================

def assistente():
    gravar_audio()
    texto = transcrever_audio("audio.wav")
    resposta = gerar_resposta(texto)
    
    print("Resposta:", resposta)
    
    sintetizar_voz(resposta)

#Executar
assistente()
```
