```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "gosto de machine learning",
    "machjine learning é incrível",
    "gosto de programar"
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

print("Vocabulário:")
print(vectorizer.get_feature_names_out())

print("\nMatriz")
print(X.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# 1. Dados de treino (Exemplos reais de um e-commerce)
corpus_treino = [
    "O produto é excelente, adorei a compra", # Positivo | Posição: 0
    "Muito bom, chegou antes do prazo",       # Positivo | Posição: 1  
    "Horrível, o produto veio quebrado",      # Negativo | Posição: 2
    "Não gostei, qualidade muito baixa",      # Negativo | Posição: 3
    "Entrega rápida e material de primeira"   # Positivo | Posição: 4
]

# Rótulo: 1 para Postivo, 0 para negativo
y_treino = [1, 1, 0, 0, 1]

# 2. Criando o Bag of Words
# O CountVectorizer faz o 'tokenization' e a contagem automaticamente
vectorizer = CountVectorizer()
X_treino_bow = vectorizer.fit_transform(corpus_treino)

# 3. Treinando um classificador simples (Naive Bayes)
# O Naive Bayes é excelente para trabalhar com contagens de palavras
modelo = MultinomialNB()
modelo.fit(X_treino_bow, y_treino)

# 4. Aplicação Real: Testando com um novo comnetário
#novo_comentario = ["O material é bom e a entrega foi rápida"]
novo_comentario = ["O produto não presta, material de qualidade"]
novo_v_bow = vectorizer.transform(novo_comentario) # Transforma o novo texto no mesmo formato

predicao = modelo.predict(novo_v_bow)
sentimento = "Positivo" if predicao[0] == 1 else "Negativo"

print(f"Comentário: {novo_comentario[0]}")
print(f"Predição do Modelo: {sentimento}")

# Extra: Ver o vocabulário criado
print("\nVocabulário (Bag of Words):")
print(f"\nTamanho Vocabulário:{len(vectorizer.get_feature_names_out())}")
print(vectorizer.get_feature_names_out())
```
