```python
import spacy

# 1. Carrega o modelo de portugues
nlp = spacy.load("pt_core_news_sm")

# 2. O texto que vamos analisar
#texto = "O professor tiago está ensinando Processamento de Linguagem Natural na Fatec em Jáu"
texto = "O aluno Américo está buscando a graxa em pó na Construmarques."

# 3. Criando o objeto 'doc' (a mágica acontece aqui)
doc = nlp(texto)

print("Texto processado com sucesso!")
```

```python
print(f"{'Token':<15} | {'Raiz (Lemma)':<15} | {'Classe gramatical'}")
print("-" *50)

for token in doc:
    print(f"{token.text:<15} | {token.lemma_:<15} | {token.pos_}")
```

```python
print("Entidades Encontradas: \n")

for ent in doc.ents:
    print(f"Entidade: {ent.text} -> Rótulo:{ent.label_}")
    
#Visualização gráfica (Exclusivo para Jupyter/Cobal)
from spacy import displacy
displacy.render(doc, style="ent",jupyter=True)
```
