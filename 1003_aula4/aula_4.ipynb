```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 1. Definimos os documento (o nosso corpus)

corpus = [
    "O gato comeu o peixe",
    "O gato dormui no sofá",
    "O cachorro correu no parque"
]

# 2. Inicializamos o vectorizer
# O parâmetro stop_words pode ser usado para ignorar
# palabras como "o", "no", etc.
vectorizer = TfidfVectorizer()

# 3. Ajustamos e transformamos os dados
tfidf_matrix = vectorizer.fit_transform(corpus)

# 4. Transformamos em um DataFrame para facilitar a visualização
df_tfidf = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out(),
    index=['Documento 1', 'Documento 2', 'Documento 3']
)

# Define para não truncar as colunas (None significa sem limite)
pd.set_option('display.max_columns', None)

# Opcional: define a largura da linha para não quebrar o texto
pd.set_option('display.width', 10000)
pd.set_option('display.precision', 5)

print(df_tfidf)
```
