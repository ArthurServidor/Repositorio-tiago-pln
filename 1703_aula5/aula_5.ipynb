```python
from gensim.models import Word2Vec

sentences = [
    ["gato", "come", "peixe"],
    ["cachorro", "come", "carne"],
    ["gato", "e", "cachorro", "são", "animais"]
]

# Treinando o modelo
# vector_size: tamanho do vetor (embedding)
# window: janela de contexto
# min_count: ignora palavras que aparecem menos que X vezes
model = Word2Vec(sentences, vector_size=2, window=2, min_count=1)

print(model.wv["gato"])
```

```python
model.wv.similarity("gato", "cachorro")
```

```python
model.wv.most_similar("gato")
```

```python
import matplotlib.pyplot as plt

words = list(model.wv.index_to_key)

for word in words:
    x, y = model.wv[word]
    plt.scatter(x, y)
    plt.text(x-0.05, y+0.01, word)
    
plt.show()
```

```python
model.wv.most_similar("gato")
```

```python
model.wv.most_similar(
    positive=['gato', 'come'],
    negative=['peixe']
)
```

```python
from gensim.models import Word2Vec
import matplotlib.pyplot as plt

sentencas = [
    ["o", "processamento", "de", "linguagem", "natural", "e", "incrivel"],
    ["aprendizado", "de", "maquina", "e", "uma", "subarea", "da", "ia"],
    ["redes", "neurais", "sao", "usadas", "em", "deep", "learning"],
    ["o", "gensim", "e", "excelente", "para", "word", "embeddings"]
]

# Treinando o modelo
# vector_size: tamanho do vetor (embedding)
# window: janela de contexto
# min_count: ignora palavras que aparecem menos que X vezes
model = Word2Vec(sentencas, vector_size=2, window=5, min_count=1)

try:
    print(model.wv["gato"])
except KeyError:
    print("esta palavra não existe no vocabulario")
```

```python
model.wv.most_similar("linguagem")
```

```python
words = list(model.wv.index_to_key)

for word in words:
    x, y = model.wv[word]
    plt.scatter(x, y)
    plt.text(x+0.01, y+0.01, word)

plt. show()
```
